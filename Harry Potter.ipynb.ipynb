# Visualizing the datset

```python
plt.figure(figsize=(25, 6))
plt1 = d3["Chapter Name"][:75].value_counts().plot(kind='bar')
plt.title('Chapter count')

plt1.set(xlabel = 'Chapters Name', ylabel='popularness of chapters')
plt.show()
```

### - Doorstep Delivery and Out of the Hat are most famous chapters
### - They are equally famous

```python
plt.figure(figsize=(25, 6))
plt1 = d3["Movie Title"].value_counts().plot(kind='bar')
plt.title('Movie analysis')
plt1.set(xlabel = 'Movie Name', ylabel='popularness of movies')
plt.show()
```
